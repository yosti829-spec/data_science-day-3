# **Step 2: Data Cleaning**

Handling Missing Values in Age Column
The missing values in the Age column are replaced with the median value. Median is used because it is not affected by extreme values.

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

Handling Missing Values in Embarked Column
The missing values in the Embarked column are filled with the mode (most frequent value).

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

Dropping Cabin Column
The Cabin column is removed because it contains many missing values and is not useful for analysis.

```python
df = df.drop(columns=['Cabin'])
```

Removing Duplicate Rows
Duplicate rows are removed to ensure data accuracy and avoid repeated information.

```python
df = df.drop_duplicates()
```

Checking the Dataset After Cleaning
This step ensures that missing values are handled and the dataset is clean.

```python
df.info()
```

# **Step 3: Data Analysis**

Survival Rate by Gender
This analysis shows the survival rate of passengers based on gender. It helps to understand whether males or females had a higher chance of survival.

```python
df.groupby('Sex')['Survived'].mean()
```

Survival Rate by Passenger Class
This shows how survival rate varies across different passenger classes (1st, 2nd, 3rd). It helps to understand the impact of social class on survival.

```python
df.groupby('Pclass')['Survived'].mean()
```

Average Age per Class
This calculates the average age of passengers in each class. It helps to analyze the age distribution among classes.

```python
df.groupby('Pclass')['Age'].mean()
```

Survival Rate by Age Group
Passengers are divided into age groups (Child, Teen, Adult, Senior). This helps to see which age group had higher survival rates.

```python
df['AgeGroup'] = pd.cut(df['Age'],
                       bins=[0, 12, 18, 60, 100],
                       labels=['Child', 'Teen', 'Adult', 'Senior'])

df.groupby('AgeGroup')['Survived'].mean()
```

# **Step 4: Filtering**

Female Passengers Who Survived
This filters the dataset to show only female passengers who survived the Titanic disaster.

```python
df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
```

Children Who Survived
This filters passengers who are considered children (Age less than 12) and survived.

```python
df[(df['Age'] < 12) & (df['Survived'] == 1)]
```

1st Class Passengers Who Survived
This filters passengers from 1st class who survived, showing those with a higher survival probability.

```python
df[(df['Pclass'] == 1) & (df['Survived'] == 1)]
```
